```python
import pandas as pd
import os
```

Merging 12 months of data into one single file

```python
df = pd.read_csv('./Data_Files/Sales_April_2019.csv')

files = [file for file in os.listdir('./Data_Files')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Data_Files/' + file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv('all_data.csv', index = False)
```

Read in updated dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

Clean up Data!

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how = 'all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data.head()
```

Q1: What was the best month for sale and how much was earned that month?

```python
# Adding month column

all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] == all_data['Month'].astype('int32')

all_data.head()
```

```python
# Adding Sales Column

all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']

all_data.head()
```

```python
results = all_data.groupby('Month').sum()
```

```python
#Ans: Best month was December with 4.613443
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks = months
plt.ylabel('Sales in USD')
plt.xlabel('Months')
plt.show()
```

Q2: What city had the highest number of sales?

```python
#Creating a Column

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_data.head(12)
```

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt


city=all_data.groupby("City").sum().reset_index().sort_values("Sales",ascending=False)

cities = all_data['City'].unique()

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size = 5)
plt.ylabel ('Sales in USD')
plt.xlabel("City Name")

plt.show()
```

Q3: What time should we display advertisements to maximize likelihood of customer's buying product?

Q4: What Products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key,value in count.most_common(10):
    print(key,value)
```
